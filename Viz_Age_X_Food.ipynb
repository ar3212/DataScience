```
import pandas as pd
import numpy as np
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
colors=sns.color_palette("tab10")
```

```
df=pd.read_csv(data)
df.head()
```

```
df.info()
```

```
df.isna().sum()
```

```
def klasifikasi(Age):
  if Age<=20: return "Teen"
  if Age>20 and Age<=30: return "Young Adult"
  if Age>30 and Age<=40: return "Mid Adult"
  if Age>40 and Age<=50: return "Old Adult"
  if Age>50: return "Elderly"

df["Gen"]=df['Age in years'].apply(klasifikasi)
```

```
df.head(10)
```

```
data=df.sort_values("Age in years",ascending=True)
data
```

```
#Beause of data with Boolean it better to drop the na 
data.dropna(inplace=True)
data.isna().sum()
```

```
import matplotlib.pyplot as plt
```

```
pd.unique(data.Gen)
```

```
plt.hist(data[data["Gen"]=="Teen"]["Age in years"],color=colors[0],label="Teen")
plt.hist(data[data["Gen"]=="Young Adult"]["Age in years"], color=colors[1], label="Young Adult")
plt.hist(data[data["Gen"]=="Mid Adult"]["Age in years"], color=colors[2], label="Mid Adult")
plt.hist(data[data["Gen"]=="Old Adult"]["Age in years"], color=colors[3], label="Old Adult")
plt.hist(data[data["Gen"]=="Elderly"]["Age in years"], color=colors[4],label="Elderly")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.figure(figsize=(1000,1000),dpi=1000)
plt.show()
```

```
data.describe().T
```

```
data.info()
```

```
food=data.drop(columns=["Age in years","Gender","Race","Home location"], axis=1)
summary=food.groupby(["Gen"]).agg("sum")
summary=summary.reindex(["Teen","Young Adult","Mid Adult","Old Adult","Elderly"])
summary["count"]=food.groupby(["Gen"]).agg("count").iloc[:,0]
summary
```

```
#We see how age ussually eat food
plt.figure(figsize=(16,16))
_=sns.heatmap(summary.drop(["count"],axis=1).T,annot=True,fmt="d")
```

```
#Elderly less frequently eat more varian food.
```

```
#Ration per Age
summary_ratio=summary.apply(lambda x:x/x['count']*100, axis=1)
summary_ratio.drop(["count"],axis=1,inplace=True)
summary_ratio.style.background_gradient(cmap=sns.light_palette("seagreen", as_cmap=True))
```
