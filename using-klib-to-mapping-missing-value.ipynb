```python
data=("../input/california-housing-prices/housing.csv")
```

```python
#Importing data and package
import klib
import pandas as pd
df=pd.read_csv(data)
df
```

```python
#Plotting Missing Value
klib.missingval_plot(df)

#in the Header or top is the general statistic of information for missing value
#in the Bottom show the distribution of missing value by black color in the dataframe
```

```python
data_clean=klib.data_cleaning(df)
```

```python
klib.corr_plot(data_clean, annot=True)
```

```python
klib.corr_plot(data_clean,split="pos", annot=True)
```

```python
klib.corr_plot(data_clean,split="neg", annot=True)
```

```python
plot=klib.corr_plot(data_clean,target="median_house_value", annot=True)
```

```python
klib.corr_mat(data_clean)
```

```python
klib.dist_plot(data_clean['median_income'])
```

```python
klib.cat_plot(df)
```

```python
#SavePlot as Pdf
plot.figure.savefig("plotcor.pdf")
```
