```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

Loading Data

```python
crash_df=sns.load_dataset('car_crashes')
crash_df.head()
```

# Continues Plot

```python
#Pair Plot
_=sns.pairplot(crash_df)
```

```python
#Dist Plot
_=sns.distplot(crash_df['speeding'], kde=False, bins=25)
```

```python
#Join Plot
_=sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
_=sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
_=sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
#KDE Plot
_=sns.kdeplot(crash_df['alcohol'])
_=sns.kdeplot(crash_df['speeding'])
```

```python
#We can categorize by using hue as sex on data tip
tips=sns.load_dataset('tips')

_=sns.pairplot(tips, hue='sex', palette='Blues')
```

```python
#Rug Plot
_=sns.rugplot(tips['tip'])
```

```python
#Styling
plt.figure(figsize=(10,20))
sns.set_style('ticks')

sns.set_context('paper', font_scale=1.4)
_=sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=True, bottom=True, right=True, top=True)
```

# Categorical Plot

```python
#Bar Plot
plt.figure(figsize=(20,10))
_=sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
#count Plot
plt.figure(figsize=(20,10))
_=sns.countplot(x='sex', data=tips, dodge=True,hue='smoker', palette='mako')
```

```python
#Box Plot
plt.figure(figsize=(20,10))
_=sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='viridis')
```

```python
#Violin Plot
plt.figure(figsize=(20,10))
_=sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='magma')
```

```python
plt.figure(figsize=(20,10))
_=sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='magma', split=True)
```

```python
#Strip Plot
plt.figure(figsize=(20,10))

_=sns.stripplot(x='day', y='total_bill', hue='sex', palette='mako',jitter=True ,data=tips, dodge=True)
```

```python
#Swarm Plot
plt.figure(figsize=(20,10))
_=sns.swarmplot(x='day', y='total_bill', data=tips, color='blue')
```

```python
#Swarm Plot x Violin Plot
plt.figure(figsize=(20,10))
_=sns.violinplot(x='day', y='total_bill', data=tips, palette='hls')
_=sns.swarmplot(x='day', y='total_bill', data=tips, color='white')
```

```python
#Test Pallete
#You can using pallete from matplotlib

plt.figure(figsize=(20,10))
sns.set_style('white')
sns.set_context('talk')

sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', palette='afmhot')
plt.legend(loc=4)

#option for loc : 1 = upper right, 2= upper left,lower left 3, lower right =4
```

```python
#Heatmap

plt.figure(figsize=(20,10))
sns.set_context('paper',font_scale=1.4)

_=sns.heatmap(crash_df.corr(), annot=True, cmap='mako')
```

```python
#Heatmap form pivot
flights=sns.load_dataset('flights')
flights.head()
```

```python
#Pivot
flights_piv=flights.pivot_table(index='month', columns='year', values='passengers')
flights_piv.head()
```

```python
  plt.figure(figsize=(20,10))
_=sns.heatmap(flights_piv, annot=True, cmap='cividis', linecolor='white', linewidth=1, fmt='g')
```

```python
#Cluster Map
iris=sns.load_dataset('iris')
species=iris.pop('species')
_=sns.clustermap(iris)
```

```python
_=sns.clustermap(flights_piv, cmap='cividis')
```

```python
#Pair Grid
iris=sns.load_dataset('iris')
iris_a=sns.PairGrid(iris, hue='species')
iris_a.map(plt.scatter)
iris_a.map_diag(plt.hist)
```

```python
iris=sns.load_dataset('iris')
iris_a=sns.PairGrid(iris, hue='species')
iris_a.map_diag(plt.hist)
#iris_a.map_offdiag(plt.scatter)

iris_a.map_upper(plt.scatter)
iris_a.map_lower(sns.kdeplot)
```

```python
plt.figure(figsize=(20,10))
iris_b=sns.PairGrid(iris,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
iris_b.map(plt.scatter)
iris_b.add_legend()
```

```python
#Facetgrid
tip=sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
tip.map(plt.scatter, 'total_bill', 'tip')
tip.add_legend()
```

```python
at_df=sns.load_dataset('attention')
at_fg=sns.FacetGrid(at_df, col='subject',col_wrap=5,height=1.5)
at_fg.map(plt.plot,'solutions','score', marker='')
```
